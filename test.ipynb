```python
def get_number(prompt="Enter a number: "):
    while True:
        try:
            return int(input(prompt))
        except ValueError:
            print("Invalid input. Please enter a valid number.")

def get_operation():
    valid_operations = ('+', '-', '*', '/', '%', '**', '//', '=', 'clear', 'clearall', 'exit')
    while True:
        symbol = input("Enter an operation (+, -, *, /, %, **, //, =, clear, clearall, exit): ")
        if symbol in valid_operations:
            return symbol
        print("Invalid operation. Please enter a valid symbol.")

user_input_history = []
output = None

while True:
    print("Current History:", user_input_history)

    # Check if last entry is an operator or the list is empty to prompt for a number
    if len(user_input_history) == 0 or isinstance(user_input_history[-1], str):
        temp = get_number()
        user_input_history.append(temp)
        if output is None:
            output = temp  # Initialize output with the first number
    else:
        # Prompt for an operation if the last entry was a number
        symbol = get_operation()
        user_input_history.append(symbol)

        if symbol in ('+', '-', '*', '/', '%', '**', '//'):
            temp = get_number("Enter another number: ")
            user_input_history.append(temp)

            # Perform the operation based on the symbol
            try:
                if symbol == '+':
                    output += temp
                elif symbol == '-':
                    output -= temp
                elif symbol == '*':
                    output *= temp
                elif symbol == '/':
                    output /= temp
                elif symbol == '%':
                    output %= temp
                elif symbol == '**':
                    output **= temp
                elif symbol == '//':
                    output //= temp
                print(f"Result after {symbol} is:", output)
            except ZeroDivisionError:
                print("Error: Division or Modulus by zero is not allowed.")
                user_input_history.pop()  # Remove invalid entry

        elif symbol == '=':
            print("Final Output:", output)
            break

        elif symbol == "clear":
            output = 0
            print("Output reset. History kept.")

        elif symbol == "clearall":
            user_input_history.clear()
            output = None
            print("All history cleared.")

        elif symbol == "exit":
            print("Exiting the calculator.")
            break
```

```python
user_input_history = []
output = None  # Initialize output to None to handle the first number properly

while True:
    print("Current History:", user_input_history)

    # Check if we need to input a number or an operation
    if len(user_input_history) == 0 or isinstance(user_input_history[-1], str):
        # Prompt for a number
        try:
            temp = int(input("Enter a number: "))
            user_input_history.append(temp)
            if output is None:
                output = temp  # Set the initial output value
        except ValueError:
            print("Invalid input. Please enter a valid number.")
            continue  # Skip to the next iteration if the input is invalid
    else:
        # Prompt for an operation
        symbol = input("Enter an operation (+, -, *, /, %, **, //, =, clear, clearall, exit): ")
        
        if symbol not in ('+', '-', '*', '/', '%', '**', '//', '=', 'clear', 'clearall', 'exit'):
            print("Invalid operation. Please enter a valid symbol.")
            continue  # Skip the rest of the loop if the operation is invalid

        user_input_history.append(symbol)  # Append the symbol to the history

        match symbol:
            case '+':
                temp = int(input("Enter another number: "))
                user_input_history.append(temp)
                output += temp
                print("Addition result:", output)

            case '-':
                temp = int(input("Enter another number: "))
                user_input_history.append(temp)
                output -= temp
                print("Subtraction result:", output)

            case '*':
                temp = int(input("Enter another number: "))
                user_input_history.append(temp)
                output *= temp
                print("Multiplication result:", output)

            case '/':
                temp = int(input("Enter another number: "))
                user_input_history.append(temp)
                if temp == 0:
                    print("Cannot divide by zero.")
                    user_input_history.pop()  # Remove the invalid number
                    continue
                output /= temp
                print("Division result:", output)

            case '%':
                temp = int(input("Enter another number: "))
                user_input_history.append(temp)
                if temp == 0:
                    print("Cannot perform modulus by zero.")
                    user_input_history.pop()
                    continue
                output %= temp
                print("Modulus result:", output)

            case '**':
                temp = int(input("Enter another number: "))
                user_input_history.append(temp)
                output **= temp
                print("Exponentiation result:", output)

            case '//':
                temp = int(input("Enter another number: "))
                user_input_history.append(temp)
                if temp == 0:
                    print("Cannot perform floor division by zero.")
                    user_input_history.pop()
                    continue
                output //= temp
                print("Floor Division result:", output)

            case '=':
                print("Final Output:", output)
                break

            case "clear":
                output = 0
                print("Output reset. History kept.")

            case "clearall":
                user_input_history.clear()
                output = None
                print("All history cleared.")

            case "exit":
                print("Exiting the calculator.")
                break
```
