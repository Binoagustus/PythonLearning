Input
1. Enter a number for an operation in calculator
2. Get a User Input of Number, following an operator, then a number in a consequential manner
3. Condition used in if condition is to check from the list, whether the last element is an operator, to get the user input as number

Output
1. output as of in a calculator
2. View the history
3. clear history

```python
#Append the temp value to history
user_history = []

def number_input():
    print("Enter another number: ")
    temp = int(input())
    user_history.append(temp)
    return temp

def add(output):
    temp = number_input()
    output += temp
    print("Addition result:", output)
    return output

def sub(output):
    temp = number_input()
    output -= temp
    print("Subtraction result:", output)
    return output

def mult(output):
    temp = number_input()
    output *= temp
    print("Multiplication result:", output)
    return output

def exponentiation(output):
    temp = number_input()
    output **= temp
    print("Exponentiation result:", output)
    return output

#This does the operations and adds the history to user_history list
def arithmetic_operations():
    output = 0
    while True:   
        # Check if we need to input a number or an operation
        if (len(user_history) == 0 or user_history[-1] in ('+', '-', '*', '/', '%', '**', '//', '=', 'exit')):
            # Prompt for a number
            try:
                print("Enter a number: ")
                temp = int(input())
                user_history.append(temp) #Append the temp value to history
                if output == 0:
                    output = temp  # Set the initial output value
            except ValueError:
                print("Invalid input. Please enter a valid number.")
                continue  # Skip to the next iteration if the input is invalid
        else:
            # Prompt for an operation
            try:
                print("Enter an operation (+, -, *, /, %, **, //, =, exit): ")
                symbol = input()
                
                if symbol not in ('+', '-', '*', '/', '%', '**', '//', '=', 'exit'):
                    print("Invalid operation. Please enter a valid symbol.")
                    continue  # Skip the rest of the loop if the operation is invalid
            except ValueError:
                print("Invalid input. Please enter a valid Operator.")
                continue

            user_history.append(symbol)  # Append the symbol to the history

            match symbol:
                case '+':
                    output = add(output)

                case '-':
                    output = sub(output)

                case '*':
                    output = mult(output)

                case '/':
                    temp = number_input()
                    if temp == 0:
                        print("Cannot divide by zero.")
                        user_history.pop()  #Removes the last element which was added to the list ,i.e 0
                        continue
                    output /= temp
                    print("Division result:", output)

                case '%':
                    temp = number_input()
                    if temp == 0:
                        print("Cannot perform modulus by zero.")
                        user_history.pop()
                        continue
                    output %= temp
                    print("Modulus result:", output)

                case '**':
                    output = exponentiation(output)

                case '//':
                    temp = number_input()
                    if temp == 0:
                        print("Cannot perform floor division by zero.")
                        user_history.pop()
                        continue
                    output //= temp
                    print("Floor Division result:", output)

                case '=':
                    print("Final Output:", output)
                    return(output, user_history)

                case "exit":
                    print("Exiting the calculator.")
                    return(output, user_history)

#User Operation
def user_input():
    history = []
    answer = 0
    while(True):
        print(''' 
            Calculator
                1. Enter Arithmetic Operation
                2. View History
                3. Clear History
                4. Exit Calculator
            Enter a number to perform operation       
            ''')
        try:
            case_input = int(input())

            match case_input:
                case 1:
                    #Setting the output and history to local variable
                    value_op = arithmetic_operations()
                    answer = value_op[0]
                    history = value_op[1]

                case 2:
                    if(len(history) == 0):
                        print("History is empty")
                    else: 
                        print("History is ", history)

                case 3:
                    history.clear()
                    print("History cleared")

                case 4:
                    break
        except ValueError:
            print("Please enter a valid number")

#Program starts here
def main():
    user_input()
                
main()
```
